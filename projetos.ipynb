```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
colunasSelecionadas = [
        'CD_GENERO', 'DS_GENERO', 
        'DS_FAIXA_ETARIA',
       'CD_GRAU_ESCOLARIDADE', 'DS_GRAU_ESCOLARIDADE']
```

```python
dadosSelecionados = dados.filter(items=colunasSelecionadas)
```

```python
dadosSelecionados.head()
```

```python
plt.figure(figsize = (30,15))
dados["DS_GRAU_ESCOLARIDADE"].hist(bins = 30, ec = "k", alpha = .6, color = "royalblue")
plt.suptitle('Grau de Escolaridade dos Eleitores',  
             ha = 'center', 
             fontsize = 50, 
             weight = 'extra bold')

plt.xlabel("Quantidade de Eleitores")
plt.ylabel("Grau de Escolaridade")
```

```python
plt.figure(figsize = (40,15))
dados['DS_FAIXA_ETARIA'].hist(bins = 30, ec = "k", alpha = .6, color = "royalblue")
plt.suptitle('Faixa Etaria dos Eleitores',  
             ha = 'center', 
             fontsize = 50, 
             weight = 'extra bold')

plt.xlabel("Quantidade de Eleitores")
plt.ylabel("Faixa Etaria")
```
